```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las funciones
def f1(x):
    return 180 - 2*x

def f2(x):
    return (160 - x) / 2

def f3(x):
    return 100 - x

# Creamos un rango de valores de x
x = np.arange(0, 100)

# Calculamos las intersecciones entre las funciones
x1 = np.min(np.roots([2, 1, -180]))
x2 = np.min(np.roots([1, 2, -160]))
x3 = np.min(np.roots([1, 1, -100]))

# Creamos un rango de valores de y
y = np.arange(0, 100)

# Graficamos las tres funciones
plt.plot(x, f1(x), label='2x + y = 180')
plt.plot(x, f2(x), label='x + 2y = 160')
plt.plot(x, f3(x), label='x + y = 100')

# Graficamos las regiones factibles
plt.fill_between(x, f1(x), 0, where=(x <= x1), color='gray', alpha=0.5)
plt.fill_between(x, f2(x), 0, where=(x <= x2), color='gray', alpha=0.5)
plt.fill_between(x, f3(x), 0, where=(x <= x3), color='gray', alpha=0.5)

# Configuramos los límites de los ejes
plt.xlim(0, 100)
plt.ylim(0, 100)

# Agregamos etiquetas a los ejes y un título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de las funciones 2x + y < 180, x + 2y < 160, x + y < 100')

# Agregamos una leyenda
plt.legend()

# Mostramos la gráfica
plt.show()
```
