```python
import matplotlib.pyplot as plt
import numpy as np

# Crear los valores de x que queremos graficar
x = np.linspace(-10, 10, 1000)

# Crear los valores de y
y = np.full_like(x, 5)

# Crear la gráfica
plt.plot(x, y)

# Definir el rango de la gráfica en el eje y
plt.ylim(-10, 10)

# Mostrar la gráfica
plt.show()
```
