```python
# Graficar la linea
import matplotlib.pyplot as plt
import numpy as np

def func(x):
    return 5 - 2*x

x = np.linspace(-5, 5, 100)
y = func(x)

fig, ax = plt.subplots()
ax.plot(x, y, '-r', label='2x + y = 5')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='upper left')

plt.show()
```
