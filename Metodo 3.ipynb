```python
import matplotlib.pyplot as plt
import numpy as np

# Definir la ecuación de la recta
def func(x):
    return 0.5*x + 0.5

# Generar valores de x
x = np.linspace(-10, 10, 1000)

# Calcular valores de y
y = func(x)

# Crear el gráfico
plt.plot(x, y)

# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de la función y = 0.5x + 0.5')

# Mostrar el gráfico
plt.show()
```
