```python
import matplotlib.pyplot as plt
import numpy as np

# Graficar las restricciones
x = np.linspace(0, 65, 100)
y1 = 65 - x
y2 = np.ones_like(x) * 20
y3 = x - 23
y4 = (3000 - 60 * x) / 24

plt.plot(x, y1, label='Total de cursos')
plt.plot(x, y2, label='Cursos ajenos a administración')
plt.plot(y3, x, label='Cursos de administración')
plt.plot(x, y4, label='Presupuesto de libros')

# Configurar el gráfico
plt.xlim([0, 65])
plt.ylim([0, 65])
plt.xlabel('Cursos de administración')
plt.ylabel('Cursos ajenos a administración')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
print('x es:',40)
print('y es:', 25)
```
