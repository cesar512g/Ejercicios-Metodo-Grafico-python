```python
import matplotlib.pyplot as plt
import numpy as np

# Creamos un rango de valores de x
x = np.arange(0, 31)

# Calculamos los valores correspondientes de y
y = (60 - 2*x) / 3

# Graficamos la ecuación
plt.plot(x, y)

# Configuramos los límites de los ejes
plt.xlim(0, 30)
plt.ylim(0, 20)

# Agregamos etiquetas a los ejes y un título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la ecuación 2x + 3y = 60')

# Mostramos la gráfica
plt.show()
```
