```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos las dos funciones
def f1(x):
    return (1/2) * x +1/2

def f2(x):
    return 3 - 2 * x

# Creamos un conjunto de valores x para graficar
x = np.linspace(-10, 10, 100)

# Evaluamos las funciones en cada valor de x
y1 = f1(x)
y2 = f2(x)

# Creamos la gráfica
plt.plot(x, y1, label='2y - 1 = 0')
plt.plot(x, y2, label='2x + y = 3')

# Agregamos etiquetas a los ejes
plt.xlabel('x')
plt.ylabel('y')

# Agregamos una leyenda
plt.legend()

# Mostramos la gráfica
plt.show()
```
