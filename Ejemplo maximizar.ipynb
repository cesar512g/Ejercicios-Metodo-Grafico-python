```python
import matplotlib.pyplot as plt
import numpy as np

# Definir las variables y la función objetivo
x = np.linspace(0, 200, 1000)
y1 = (30e6 - 1e6*x)/3e5
y2 = (30e6 - 3e5*x)/1.2e6

# Graficar las restricciones
plt.plot(x, y1, label='Fútbol')
plt.plot(x, y2, label='Básquetbol')

# Graficar la región factible
y3 = np.maximum(np.minimum(y1, y2), 0)
plt.fill_between(x, y3, 0, where=(y3>0), alpha=0.2)

# Definir el punto extremo de la región factible
x_opt = 200/3
y_opt = 100


# Configurar el gráfico
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.xlabel('Equipos de fútbol')
plt.ylabel('Equipos de básquetbol')
plt.legend()

# Mostrar el gráfico
plt.show()

# Imprimir la solución
print('El número máximo de equipos de fútbol es:',12)
print('El número máximo de equipos de básquetbol es:', 26)
```
